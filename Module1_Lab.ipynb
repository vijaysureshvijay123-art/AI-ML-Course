```python
 import numpy as np
 import numpy.linalg as la
 a = np.array([[4,5],[2,3]])
 #determinent function
 m=(la.det(a))
 print(m)
```

```python
import numpy as np

matrix = np.array([[1, 2],
                   [3, 4]])

rank = np.linalg.matrix_rank(matrix)


print(f"{rank}")
```
